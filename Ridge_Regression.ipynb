First of all we will be implementing the multiple linear regression on sythethic dataset  after that we will be using Ridge regression and will compare the accuracy of both the models

# Generate a Complex Synthetic Dataset

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

# Generate synthetic data
np.random.seed(0)
n_samples = 100
n_features = 5

X = np.random.randn(n_samples, n_features)
coefficients = np.array([1.5, -2., 3., 0.5, -1.])
y = np.dot(X, coefficients) + np.random.normal(scale=2, size=n_samples)

# Create a DataFrame for better visualization
data = pd.DataFrame(X, columns=[f'Feature_{i}' for i in range(1, n_features+1)])
data['Target'] = y

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# Implement Multiple Linear Regression on the Dataset

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Train the Multiple Linear Regression model
mlr = LinearRegression()
mlr.fit(X_train, y_train)

# Predict and evaluate
y_pred_mlr = mlr.predict(X_test)
mse_mlr = mean_squared_error(y_test, y_pred_mlr)
r2_mlr = r2_score(y_test, y_pred_mlr)

mse_mlr, r2_mlr
```

# Implement Ridge Regression on the Dataset

```python
from sklearn.linear_model import Ridge

# Train the Ridge Regression model
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)

# Predict and evaluate
y_pred_ridge = ridge.predict(X_test)
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)

mse_ridge, r2_ridge
```

```python
import matplotlib.pyplot as plt

plt.subplot(1, 2, 1)
plt.scatter(y_test, y_pred_mlr, edgecolor='k', alpha=0.7, label='Predictions')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', lw=2)
plt.title('Multiple Linear Regression')
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(y_test, y_pred_ridge, edgecolor='k', alpha=0.7, label='Predictions')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', lw=2)
plt.title('Ridge Regression')
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.legend()

plt.show()
```
